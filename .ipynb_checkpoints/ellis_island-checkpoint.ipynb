```python
import pandas as pd
import requests
import io
import os
import seaborn as sns
import geocoder

sns.set_style()
%matplotlib inline
```

```python
#create_structure.py

#basic setup of a project directory

paths = ['./data','./data/raw','./data/processed','./notebooks']
for path in paths:
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
#download_data.py

#go out and get our data

DATA_URL = "https://raw.githubusercontent.com/hopperrr/ellis-immigration-by-ship/gh-pages/data/trips.tsv"
data = requests.get(DATA_URL)
stream = io.StringIO(data.text)
df = pd.read_csv(stream, sep='\t')
df.to_csv('./data/raw/ellis_arrivals.csv', index=False)
```

```python
#process_data.py

df = pd.read_csv('./data/raw/ellis_arrivals.csv')
df.head()
#data doesn't look too bad to start - we'll need to transform the date field into a data and the count into an integer.
(df.size - (df.size - df.dropna().size)) / df.size
#there's a few NA values, but they make up a small portion of the data so we'll drop them for now
df.dropna(inplace=True)

df['Full_date'] = pd.to_datetime(df['Full_date'], format='%m/%d/%Y')
df['Year'] = df['Full_date'].dt.year
# Exact date results in a lot of data points over 100 years - we'll extract the year out to reduce granularity
df['Count'] = df['Count'].astype(int)
```

```python
df.groupby('Year').sum().plot(kind='bar')
# at a first glance, we can see a few clear trends - a steady increase in arrivals in the early 1900's, then a sharp drop
# during the war years, then a retrun to pre-war levels.

# a few thigns to look at:
# why the drop in 1908? Data issue, or something else?
# it looks like our data is mostly missing after 1924 - we'll drop years 1925 and on to clean that up.
df = df[df['Year'] < 1925]
```

```python
df.groupby('Year').sum().plot(kind='bar')
```

```python
# Let's take a look at traffic by ship - were there some ships that carried more passengers?

ship_count = df.groupby('Ship').agg({'Count':sum}).sort_values(by='Count',ascending=False)
ship_count['Total'] = ship_count['Count'].sum()
ship_count['perc_of_total'] = ship_count['Count'] / ship_count['Total']
ship_count.head()
```
